# Step #1: Explore and prep data

```python
import pandas as pd

# Pitching
df = pd.read_csv('ncaa_pitchingQualifiedCSV/PitchingDraftTable.csv', delimiter=',')
```

```python
# list and drop columns that are less related to the target
cols_to_drop = ['name', 'playerid', 'mlbamid', 'team', 'Acronym', 'Full Team Name', 'Drafted By', 'Drafted From']
df = df.drop(columns=cols_to_drop)
print(df.columns.tolist())

# convert the target to numerical values
df['Drafted?'] = df['Drafted?'].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Drafted?'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns='Drafted?')
y = df['Drafted?']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=8)
```

# Step #2: Build a pipeline of training

```python
from sklearn.pipeline import Pipeline
from category_encoders.target_encoder import TargetEncoder
from xgboost import XGBClassifier

estimators = [
    ('encoder', TargetEncoder()),
    ('clf', XGBClassifier(random_state=8)) # can customize objective function with the objective parameter
]
pipe = Pipeline(steps=estimators)
pipe
```

# Step #3: Set up hyperparameter tuning

```python
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer

search_space = {
    'clf__max_depth': Integer(2,8),
    'clf__learning_rate': Real(0.001, 1.0, prior='log-uniform'),
    'clf__subsample': Real(0.5, 1.0),
    'clf__colsample_bytree': Real(0.5, 1.0),
    'clf__colsample_bylevel': Real(0.5, 1.0),
    'clf__colsample_bynode' : Real(0.5, 1.0),
    'clf__reg_alpha': Real(0.0, 10.0),
    'clf__reg_lambda': Real(0.0, 10.0),
    'clf__gamma': Real(0.0, 10.0)
}

opt = BayesSearchCV(pipe, search_space, cv=3, n_iter=10, scoring='roc_auc', random_state=8) 
# in reality, you may consider setting cv and n_iter to higher values
```

# Step #4: Train the XGBoost model

```python
opt.fit(X_train, y_train)
```

# Step #5: Evaluate the model and make predictions

```python
opt.best_estimator_
```

```python
opt.best_score_
```

```python
opt.score(X_test, y_test)
```

```python
opt.predict(X_test)
```

```python
opt.predict_proba(X_test)
```
